## 2. Monte Carlo simulations: Particle Physics (10 points)

High-energy cosmic-ray muons travel through matter and lose energy approximately linearly with the depth they traverse. Their final measured energy, $E_{\rm f}$, after passing through a distance $x$ of material can be related to their initial energy, $E_0$, through:

$$ E_{\rm f} = E_0 - \alpha x $$

where $\alpha$ is the average energy loss rate in $\rm MeV/cm$. Since muons have a distribution of initial energies, we can estimate their original energy spectrum by measuring the final energies of those that reach a detector. However, selection effects associated with the detectorâ€™s energy threshold can bias these measurements: only muons with sufficiently large $E_{\rm f}$ can be detected. Muons that lose too much energy before reaching the detector will not be observed, making the detected sample biased towards higher initial energies.

**Therefore, to understand selection bias, we want to simulate this effect using a Monte Carlo simulation.**

The purpose of this problem is to determine the bias as a function of depth for a sample of muons via a Monte Carlo calculation. To set up your simulation, assume that:

- the energy loss rate is $\alpha = 2.0\,\rm MeV/cm$.
