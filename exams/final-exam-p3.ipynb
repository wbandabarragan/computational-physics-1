## 3. Monte Carlo simulations: Bacterial Chemotaxis, Motility and Detection Bias (10 points)

Many species of bacteria move through fluids using a combination of straight runs and random changes in direction called *tumbles*. This process can be modeled as a stochastic random walk and is widely used in biophysics to study bacterial transport and chemotaxis.

In a simple 1D model, the displacement of a bacterium after a time interval $\Delta t$ can be written as:

$$x_{i+1} = x_i + v(x_i)\Delta t + \eta_i\sqrt{\Delta t}$$

where $v(x_i)$ is the chemotactic drift velocity generated by the nutrient concentration gradient, and $\eta_i$ is a Gaussian random variable representing stochastic fluctuations in the bacterial motion.

The nutrient concentration profile can be approximated by:

$$C(x)=C_0+\alpha x^2$$

where $C_0$ is the initial concentration, $\alpha$ is the quadratic concentration coefficient of the nutrient field.

Assume that the chemotactic drift velocity is proportional to the concentration gradient:

$$v(x_i)=\chi\frac{dC}{dx}\bigg|_i$$

where $\chi$ is a chemotactic sensitivity constant.


The stochastic term can be modeled as:

$$\eta_i = \sigma\,\xi_i$$

where:

$$\xi_i \sim \mathcal{N}(0,1)$$

The above stochastic term represents bacterial motility. The random tumbles and thermal fluctuations cause the bacterium to wobble around the deterministic chemotactic motion.


**Therefore, to understand stochastic bacterial motion, chemotaxis and detection bias, we want to simulate this process using a Monte Carlo simulation.**

The purpose of this problem is to determine how random fluctuations affect bacterial transport and microscope detection. To set up your simulation, assume that:
