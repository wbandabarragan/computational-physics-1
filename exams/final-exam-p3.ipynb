## 2. Monte Carlo simulations: Bacterial Chemotaxis, Motility and Detection Bias (10 points)

Many species of bacteria move through fluids using a combination of straight runs and random changes in direction called *tumbles*. This process can be modeled as a stochastic random walk and is widely used in biophysics to study bacterial transport and chemotaxis.

In a simple 1D model, the displacement of a bacterium after a time interval $\Delta t$ can be written as:

$$x_{i+1} = x_i + v \Delta t + \eta$$

where $v$ is the average drift velocity (the deterministic drift $v \Delta t$ represents chemotaxis), and $\eta$ is a Gaussian random variable representing stochastic fluctuations in the bacterial motion:

$$\eta \sim \mathcal{N}(0,\sigma^2)$$

The above term $\eta$ represents motility (the random tumbles and thermal fluctuations cause the bacterium to wobble around that deterministic path).

The bacterial concentration profile in a nutrient gradient can be approximated by:

$$C(x)=C_0 e^{-|x|/L}$$

where $C_0$ is the initial concentration and $L$ is the characteristic decay length of the nutrient field.

**Therefore, to understand stochastic bacterial motion, chemotaxis and detection bias, we want to simulate this process using a Monte Carlo simulation.**

The purpose of this problem is to determine how random fluctuations affect bacterial transport and concentration measurements. To set up your simulation, assume that:

- the nutrient decay length is $L = 200\,\mu{\rm m}$

- your microscope can only detect bacteria within the range $|x| \leq 400\,\mu{\rm m}$.
