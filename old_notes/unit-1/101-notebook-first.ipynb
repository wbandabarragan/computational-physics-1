## 5) Now, let us do some fancy plots, e.g. for equation:

$z = \sin(\sqrt{x^2+y^2})$

```python
# Let's define a function:

def f(x, y):  # Which we can call later on
    return np.sin(np.sqrt(x ** 2 + y ** 2)) # Must be indented

x = np.linspace(-10, 10, 30)  # Define x coordinates
y = np.linspace(-10, 10, 30)  # Define y coordinates

X, Y = np.meshgrid(x, y)  #Create the grid, i.e., from vector coordinates, it generates a matrix of coordinates

Z = f(X, Y) # Call function and get Z
```

```python
# Now we do some plotting:
fig = plt.figure()

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='Blues') # Creates 'n=100' contours at different Z to make the surface

ax.set_xlabel('x')  # Customised labels for axes
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# Now we do some plotting:
fig = plt.figure()

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 10, cmap='Blues') # Creates 'n=10' contours at different Z to make the surface

ax.set_xlabel('x')  # Customised labels for axes
ax.set_ylabel('y')
ax.set_zlabel('z')
```
