```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Image input:

```python
gal1 = fits.open("./NGC_3184_NA_MOM0_THINGS.FITS")

print(type(gal1))
```

```python
# Look at the header

head_gal1 = gal1[0].header

#print(head_gal1)

keys_gal1 = gal1[0].header.keys

print(keys_gal1)
```

```python
# Look at the data

data_gal1 = gal1[0].data

print(type(data_gal1), data_gal1.shape)
```

```python
# Get data into an image
imag_gal1 = gal1[0].data[0,0,:,:]

print(imag_gal1.shape)
```

### 2. Galaxy Imaging:

### 3. Isolate the galaxy: thresholding (pixel histogram)

### 4. Binary map (1 -> galaxy /0 -> background)

```python
# Apply thresholding to the cleaned image 

bin_gal1_2d = np.where(clean_gal1_2d > 0., 1., 0.)

print(bin_gal1_2d.shape, np.unique(bin_gal1_2d))
```

```python
# See the image: Fig Env.

plt.figure(figsize=(5,4))

surf = plt.imshow(bin_gal1_2d, cmap = "inferno_r")

plt.colorbar(surf)

plt.show()
```

### 5a. Calculating the galaxy diameter using extrema indices:

```python
ind_gal1 = np.where(bin_gal1_2d == 1)

print("Y indices are: ", ind_gal1[0])

print("X indices are: ", ind_gal1[1])
```

```python
# Get the min/max indices (extrema)
min_x = np.min(ind_gal1[1])
max_x = np.max(ind_gal1[1])

min_y = np.min(ind_gal1[0])
max_y = np.max(ind_gal1[0])

print("X extrema: ", min_x, max_x)
print("Y extrema: ", min_y, max_y)
```

```python
# Calculate the diameter by averaging
dx = max_x - min_x
dy = max_y - min_y

diam_gal1 = np.mean([dx, dy]) # in pixels

print("The galaxy diameter in [pixels] is: ", diam_gal1)
```

#### In physical units:

Trigonometry/Research the distance to the galaxy.

We want to use the size of the galaxy on the sky (CDELT2 = 4.166666768E-04 in degrees).

```python
# Pixel resolution ()
pix_res = gal1[0].header["CDELT2"] # in degrees

print("The size of the galaxy on the sky in [degrees] is: ", pix_res)

pix_res_rad = pix_res*np.pi/180.
```

$$\tan{\alpha}=\frac{d}{D}$$

```python
# Distance to the galaxy from NED 
dist_gal1 = 12.1e3 # in kpc

# Calculate the diameter:
diam_gal1_phys = dist_gal1*diam_gal1*np.tan(pix_res_rad)

# Print the result:

print("Diameter in [kpc] of the galaxy is:", diam_gal1_phys)
```

### 5b. Calculating the galaxy diameter using the centroid:


https://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops

```python
from skimage import filters # For thresholding
from skimage.measure import regionprops # For calculating the centroid
```

```python
# Get the centre of mass of the original image:

# 1. Get a threshold:
th_value = filters.threshold_otsu(imag_gal1)

print(th_value)

# 2. Labeled foreground (galaxy itself) -> binary map

labeled_gal1 = (imag_gal1 > th_value).astype(int)

print(np.unique(labeled_gal1))

# 3. Image features /properties

prop_gal1 = regionprops(labeled_gal1,imag_gal1)

print(prop_gal1[0].area)
print(prop_gal1[0].centroid)

cm_gal1 = np.array(prop_gal1[0].centroid)

print(cm_gal1)
```

```python
# See the image: Fig Env.

plt.figure(figsize=(5,4))

surf = plt.imshow(imag_gal1, cmap = "inferno_r")

plt.scatter(cm_gal1[1], cm_gal1[0], marker = "o", color = "blue")

plt.colorbar(surf)

plt.show()
```

```python
# 4. Diameter estimate

print(prop_gal1[0].axis_major_length)
print(prop_gal1[0].axis_minor_length)


dx_v2 = prop_gal1[0].axis_major_length
dy_v2 = prop_gal1[0].axis_minor_length
```

```python
diam_gal1_v2 = np.mean([dx_v2, dy_v2]) # in pixels

print("The galaxy diameter in [pixels] is: ", diam_gal1_v2)
```

```python
# Distance to the galaxy from NED 
dist_gal1 = 12.1e3 # in kpc

# Calculate the diameter:
diam_gal1_phys_v2 = dist_gal1*diam_gal1_v2*np.tan(pix_res_rad)

# Print the result:

print("Diameter in [kpc] of the galaxy is:", diam_gal1_phys_v2)
```

```python
# Circle coordinates

theta = np.linspace(0., 2*np.pi, 360)

# Radius 
rad_gal1 = diam_gal1_v2/2.

# Coordinates of the circles
x_circ = rad_gal1*np.cos(theta) + cm_gal1[1]
y_circ = rad_gal1*np.sin(theta) + cm_gal1[0]
```

```python
# See the image: Fig Env.

plt.figure(figsize=(5,4))

surf = plt.imshow(imag_gal1, cmap = "inferno_r")

plt.scatter(cm_gal1[1], cm_gal1[0], marker = "o", color = "blue")

plt.plot(x_circ, y_circ)

plt.colorbar(surf)

plt.show()
```
